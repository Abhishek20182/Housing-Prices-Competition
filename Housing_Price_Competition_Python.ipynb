```python
#import python library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_submission = pd.read_csv('sample_submission.csv')
```

```python
#create y
y = df_train.SalePrice
```

```python
#create x
df_train_features = ['LotArea', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = df_train[df_train_features]
```

```python
X.info()
```

```python
#split into training and validation data 
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
#model
df_model = RandomForestRegressor(random_state=1)
```

```python
#fit the model
df_model.fit(train_X, train_y)
```

```python
df_val_prediction = df_model.predict(val_X)
```

```python
#MAE(mean absolute error)
df_val_mae = mean_absolute_error(df_val_prediction, val_y)
```

```python
"{}".format(df_val_mae)
```

```python
#model for full data 
df_model_full_data = RandomForestRegressor(random_state=1)
```

```python
#fit model
df_model_full_data.fit(X, y)
```

```python
#features
test_X = df_test[df_train_features]
```

```python
test_predict = df_model_full_data.predict(test_X)
```

```python
#create submission.csv
output = pd.DataFrame({'Id': df_test.Id,
                       'SalePrice': test_predict})
output.to_csv('submission1.csv', index=False)
```
